# SIR with Influx

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

## Defining out model

def model(y,t,b,g,d):
    S,I = y
    return [-b*S*I + d,b*S*I-g*I]

b = 10
g = 4
d = 1

t = np.linspace(0,20,1000)
plt.figure(figsize=[10,5])
plt.title("β="+str(b)+", $\gamma$="+str(g)+", d="+str(d),fontsize=16)

ys = odeint(model,[1,2],t, args=(b,g,d))
plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[2,1],t, args=(b,g,d))
plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[.4,.5],t, args=(b,g,d))
plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[.1,.1],t, args=(b,g,d))
plt.plot(ys[:,0],ys[:,1],color='C0')

plt.xlabel("S - Susceptible", fontsize = 15)
plt.ylabel("I - Infected", fontsize = 15)
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

plt.plot([x_min,x_max],[0,0],color='C1', label="Nullcline I = 0")  # First Nullcline
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()
plt.plot([g/b,g/b],[y_min,y_max],color='C1',label="Nullcline S = $\gamma$/β")  # First Nullcline

S = np.linspace(x_min,x_max)
plt.plot(S,d/(b*S),color='C2',label="Nullcline d = SIβ")  # First Nullcline

plt.xlim([x_min, x_max])
plt.ylim([y_min, y_max])

plt.legend(fontsize=12)
```

## Preditor Prey

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

## Defining out model

def model(y,t,a,b,c1,c2):
    X,Y = y
    return [a*X-c1*X*Y,c2*X*Y - b*Y]

b = 1
g = 1
d = 1

t = np.linspace(0,20,1000)
plt.figure(figsize=[8,8])
plt.title("β="+str(b)+", $\gamma$="+str(g)+", d="+str(d),fontsize=16)

ys = odeint(model,[1,2],t, args=(b,g,d,1))
plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[1,3],t, args=(b,g,d,1))
plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[1,4],t, args=(b,g,d,1))
plt.plot(ys[:,0],ys[:,1],color='C0')

plt.xlabel("X - Prey",fontsize=15)
plt.ylabel("Y - Preditors",fontsize=15)

x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

plt.plot([1/1,1/1],[y_min,y_max],color='C1', label="Nullcline I = 0")  # First Nullcline
plt.plot([x_min,x_max],[1/1,1/1],color='C1', label="Nullcline I = 0")  # First Nullcline

plt.plot([0,0],[y_min,y_max],color='C1', label="Nullcline I = 0")  # First Nullcline
plt.plot([x_min,x_max],[0,0],color='C1', label="Nullcline I = 0")  # First Nullcline
```

## Pred - Prey - DDT

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def PPmodel(Y, t, a,b,c1,c2):
    x, y = Y
    dydx = [a*x - c1*x*y, c2*x*y - b*y]
    return dydx

t = np.linspace(0,20,500)  # Max around 20

Y0 = [200,  80]

## Good parameters: 
#     [1,.5,.01,.005]     - Oscillating Popilation
#     [1,.5,.01,.001]     - Oscillating Popilation w/few preditors
#     [1,.5,.1,.001]      - Apparent Preditor Collapse
#     [.01,.5,.1,.001]    - Apparent Prey Collapse

a,b,c1,c2 = [1,.5,.01,.005]

Y_sol = odeint(PPmodel,Y0,t,args=(a-.1,b,c1,c2))
Y_solO = odeint(PPmodel,Y0,t,args=(a,b,c1,c2))

#Creates two subplots and unpacks the output array immediately
X1 = Y_sol[:,0]
X2 = Y_sol[:,1]

plt.figure(figsize=(12,8))

xs = (b+.1)/c2
ys = (a-.1)/c1


plt.plot(t,X2,label='y DDT',color='r')
plt.plot(t,ys*np.ones(500),'--',color='r',label='y stable')


X1 = Y_solO[:,0]
X2 = Y_solO[:,1]
xs = b/c2
ys = a/c1

plt.plot(t,X2,label='y',color='y')
plt.plot(t,ys*np.ones(500),'--',color='y',label='y stable')

plt.xlabel('Time t')
plt.ylabel('Population')

plt.title("Pred")

plt.legend(loc='center right')
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def PPmodel(Y, t, a,b,c1,c2):
    x, y = Y
    dydx = [a*x*(1-x/100) - c1*x*y, c2*x*y - b*y]
    return dydx

t = np.linspace(0,20,500)  # Max around 20

Y0 = [200,  80]

## Good parameters: 
#     [1,.5,.01,.005]     - Oscillating Popilation
#     [1,.5,.01,.001]     - Oscillating Popilation w/few preditors
#     [1,.5,.1,.001]      - Apparent Preditor Collapse
#     [.01,.5,.1,.001]    - Apparent Prey Collapse

a,b,c1,c2 = [1,.5,.01,.005]

Y_sol = odeint(PPmodel,Y0,t,args=(a-.7,b,c1,c2))
Y_solO = odeint(PPmodel,Y0,t,args=(a,b,c1,c2))

#Creates two subplots and unpacks the output array immediately
X1 = Y_sol[:,0]
X2 = Y_sol[:,1]

plt.figure(figsize=(12,8))

xs = (b)/c2
ys = (a-.7)/c1


plt.plot(t,X1,label='x DDT',color='c')
plt.plot(t,xs*np.ones(500),'--',color='c',label='x stable')


X1 = Y_solO[:,0]
X2 = Y_solO[:,1]
xs = b/c2
ys = a/c1


plt.plot(t,X1,label='x',color='y')
plt.plot(t,xs*np.ones(500),'--',color='y',label='x stable')

plt.xlabel('Time t')
plt.ylabel('Population')

plt.legend(loc='center right')
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

## Defining out model

def model(y,t,beta,b,alpha,a):
    S,I = y
    return [b*(S+I) - a*S -beta*S*I,beta*S*I-(alpha+a)*I]

beta=0.7
b=.305
alpha=3
a=0.105+.15

t = np.linspace(0,20,1000)
plt.figure(figsize=[10,5])
#plt.title("β="+str(b)+", $\gamma$="+str(g)+", d="+str(d),fontsize=16)

ys = odeint(model,[1,2],t, args=(beta,b,alpha,a))
plt.plot(ys[:,0],ys[:,1],color='C0')
plt.plot(ys[0,0],ys[0,1],'o',color='C3')

ys = odeint(model,[2,1],t, args=(beta,b,alpha,a))
plt.plot(ys[:,0],ys[:,1],color='C0')
plt.plot(ys[0,0],ys[0,1],'o',color='C3')

ys = odeint(model,[.4,.5],t, args=(beta,b,alpha,a))
#plt.plot(ys[:,0],ys[:,1],color='C0')

ys = odeint(model,[.1,.1],t, args=(beta,b,alpha,a))
#plt.plot(ys[:,0],ys[:,1],color='C0')

plt.xlabel("S - Susceptible", fontsize = 15)
plt.ylabel("I - Infected", fontsize = 15)
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()

#plt.plot([x_min,x_max],[0,0],color='C1', label="Nullcline I = 0")  # First Nullcline
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()
plt.plot([(alpha+a)/beta,(alpha+a)/beta],[y_min,y_max],color='C1',label="Nullcline S = $\gamma$/β")  # First Nullcline

S = np.linspace(x_min,x_max,1000)
plt.plot(S,(a-b)*S/(b-beta*S),color='C2',label="Nullcline d = SIβ")  # First Nullcline

plt.xlim([x_min, x_max])
plt.ylim([y_min, y_max])

plt.legend(fontsize=12)
```
