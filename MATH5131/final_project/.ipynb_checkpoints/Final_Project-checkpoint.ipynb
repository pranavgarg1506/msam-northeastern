```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
def model(P,t,a,b,P1,P2,C1,C2):
    X,Y = P
    dXdt= (a-P1)*X-C1*Y*X
    dYdt = C2*X*Y-(b+P2)*Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
# Here prey=X and predator=Y
    
```

```python
X0= 200
Y0= 80
a,b,P1,P2,C1,C2=[1,0.5,0.1,0.1,0.01,0.005]
t = np.linspace(0,50,100)

f, ax = plt.subplots(1,2, figsize = [16,8])
sol = odeint(model,[X0,Y0],t, args=(a,b,P1,P2,C1,C2))
ax[0].plot(t,sol[:,0],'m',label="(Sardines)")
ax[0].plot(t,sol[:,1],'y',label="(Starlings)")
ax[1].plot(sol[:,0],sol[:,1],label="(sardines vs starlings)")
ax[0].legend()
ax[1].legend()

```

```python
def logistic_prey_growth_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K = params
    dXdt= a * X * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
    
```

```python
def logistic_prey_growth_with_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K = params
    dXdt= a * X * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```

```python
def logistic_prey_growth_with_alley_effect_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K, A = params
    dXdt= a * X * (X / A - 1) * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
```

```python
def logistic_prey_growth_with_weak_alley_effect_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K, A = params
    dXdt= a * X * (X / (X + A)) * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
```

```python
def logistic_prey_growth_with_alley_effect_and_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K, A = params
    dXdt= a * X * (X / A - 1) * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```

```python
def logistic_prey_growth_with_weak_alley_effect_and_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K, A = params
    dXdt= a * X * (X / (X + A)) * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```

```python
X0 = 200
Y0 = 80
t = np.linspace(0, 50, 10000)

# a,b,P1,P2,C1,C2=[1,0.5,0.1,0.1,0.01,0.005]
args_list = [[0, [200, 80], [1, 0.5, 0.01, 0.005, 800], "logistic_prey_growth_model"], \
            [1, [200, 80], [2.65, 1.06, 0.02, 0.02 * 0.215, 0.5, 0.5, 900], "logistic_prey_growth_with_pesticide_model"], \
            [2, [260, 70], [2.65, 1.06, 0.02, 0.02 * 0.215, 900, 150], "logistic_prey_growth_with_alley_effect_model"], \
            [3, [260, 70], [2.65, 1.06, 0.02, 0.02 * 0.215, 900, 150], "logistic_prey_growth_with_weak_alley_effect_model"], \
            [4, [200, 80], [1, 0.5, 0.01, 0.005, 0.1, 0.1, 800, 300], "logistic_prey_growth_with_alley_effect_and_pesticide_model"], \
            [5, [260, 70], [2.65, 1.06, 0.02, 0.02 * 0.215, 0.3, 0.3, 900, 150], "logistic_prey_growth_with_weak_alley_effect_and_pesticide_model"]];
            # [1, [100, 80], [1, 0.5, 0.01, 0.005, 300]], [1, [100, 80], [1, 0.5, 0.01, 0.005, 100]], ];

def plot_model(args, plt):
    model_number, init_val, params, title = args[0], args[1], args[2], args[3]
    if model_number == 0:
        sol = odeint(logistic_prey_growth_model, init_val, t, args = (params,))
    elif model_number == 1:
        sol = odeint(logistic_prey_growth_with_pesticide_model, init_val, t, args = (params,))
    elif model_number == 2:
        sol = odeint(logistic_prey_growth_with_alley_effect_model, init_val, t, args = (params,))
    elif model_number == 3:
        sol = odeint(logistic_prey_growth_with_weak_alley_effect_model, init_val, t, args = (params,))
    elif model_number == 4:
        sol = odeint(logistic_prey_growth_with_alley_effect_and_pesticide_model, init_val, t, args = (params,))
    elif model_number == 5:
        sol = odeint(logistic_prey_growth_with_weak_alley_effect_and_pesticide_model, init_val, t, args = (params,))
    fig, ax = plt.subplots(1, 2, figsize = (6, 3))
    fig.suptitle(title)
    # ax[0].plot(t, , 'm', label = "X(t)")
    ax[0].plot(t, sol[:, 0], 'm', label = "X(t)")
    ax[0].plot(t, sol[:, 1], 'y', label = "Y(t)")
    ax[1].plot(sol[:, 0], sol[:, 1], label = "X(t) vs Y(t)")
    ax[0].legend()
    ax[1].legend()
    # ax[2].legend()


# plot_model(args_list[0], plt)
# plot_model(args_list[1], plt)
# plot_model(args_list[2], plt)
plot_model(args_list[1], plt)
plot_model(args_list[3], plt)
plot_model(args_list[5], plt)
```
