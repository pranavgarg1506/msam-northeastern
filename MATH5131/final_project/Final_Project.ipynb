```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
def model(P,t,a,b,P1,P2,C1,C2):
    X,Y = P
    dXdt= (a-P1)*X-C1*Y*X
    dYdt = C2*X*Y-(b+P2)*Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
# Here prey=X and predator=Y
    
```

```python
X0= 200
Y0= 80
a,b,P1,P2,C1,C2=[1,0.5,0.1,0.1,0.01,0.005]
t = np.linspace(0,50,100)

f, ax = plt.subplots(1,2, figsize = [16,8])
sol = odeint(model,[X0,Y0],t, args=(a,b,P1,P2,C1,C2))
ax[0].plot(t,sol[:,0],'m',label="(Sardines)")
ax[0].plot(t,sol[:,1],'y',label="(Starlings)")
ax[1].plot(sol[:,0],sol[:,1],label="(sardines vs starlings)")
ax[0].legend()
ax[1].legend()

```

```python
def logistic_prey_growth_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K = params
    dXdt= a * X * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
    
```

```python
def logistic_prey_growth_with_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K = params
    dXdt= a * X * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```

```python
def logistic_prey_growth_with_alley_effect_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K, A = params
    dXdt= a * X * (X / A - 1) * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
```

```python
def logistic_prey_growth_with_weak_alley_effect_model(P, t, params):
    X, Y = P
    a, b, C1, C2, K, A = params
    dXdt= a * X * (X / (X + A)) * (1 - X / K) - C1 * Y * X
    dYdt = C2 * X * Y - b * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
```

```python
def logistic_prey_growth_with_alley_effect_and_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K, A = params
    dXdt= a * X * (X / A - 1) * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```

```python
def logistic_prey_growth_with_weak_alley_effect_and_pesticide_model(P, t, params):
    X, Y = P
    a, b, C1, C2, P1, P2, K, A = params
    dXdt= a * X * (X / (X + A)) * (1 - X / K) - P1 * X - C1 * Y * X
    dYdt = C2 * X * Y - (b + P2) * Y
   
    dPdt = [dXdt, dYdt]
    return dPdt
#Here prey=X and predator=Y
```
